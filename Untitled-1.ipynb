```python
import numpy as np
import pandas as pd
from scipy.interpolate import CubicSpline
from scipy.optimize import brentq

# 1. Example data: Suppose we have 5 points in (delta, implied vol).
delta_points = np.array([0.10, 0.25, 0.50, 0.75, 0.90])
implied_vols = np.array([8.26, 8.17, 8.22, 8.48, 8.75])  # in % say

# Convert vols to decimals for BS formulas
implied_vols = implied_vols / 100.0

# We also need: spot or forward, interest rates, etc.
S0 = 1.05        # e.g. EUR/USD spot
r_dom = 0.02     # domestic rate
r_for = 0.01     # foreign rate
tau = 1/12       # 1 month as fraction of year
F0 = S0 * np.exp((r_dom - r_for)*tau)  # approximate forward

# 2. Build a spline in (delta, sigma)-space
cs = CubicSpline(delta_points, implied_vols, bc_type='clamped')

def black_scholes_delta_call(S, K, r_d, r_f, sigma, T):
    """
    Returns call delta under Blackâ€“Scholes in FX context:
    delta = e^{-r_f*T} * N(d1),  (domestic numeraire)
    where
      d1 = [ln(S/K) + (r_d - r_f + sigma^2/2)*T] / (sigma*sqrt(T)).
    """
    from math import log, sqrt, exp
    from scipy.stats import norm
    
    d1 = (np.log(S/K) + (r_d - r_f + 0.5 * sigma**2)*T) / (sigma * np.sqrt(T))
    delta_call = np.exp(-r_f*T) * norm.cdf(d1)
    return delta_call

def delta_to_sigma(delta_target, K, guess_sigma=0.1):
    """
    Solve for sigma such that black_scholes_delta_call(...) = delta_target.
    Then plug that sigma into the spline cs(delta).
    We'll need a root-finding approach for the fixed-point:
      sigma == cs( delta_call ).
    """
    # The function whose root we want is: f(sigma) = sigma - cs(delta_call).
    def f(sigma):
        delta_call = black_scholes_delta_call(S0, K, r_dom, r_for, sigma, tau)
        return sigma - cs(delta_call)

    # Use a bracket for brentq. We'll assume volatility won't exceed [0.0001,2]
    try:
        sol = brentq(f, 1e-4, 2.0)
    except:
        sol = np.nan
    return sol

def black_scholes_call(S, K, r_d, r_f, sigma, T):
    """
    Returns call price under Black-Scholes (domestic discounting).
    """
    from math import log, sqrt, exp
    from scipy.stats import norm
    
    d1 = (np.log(S/K) + (r_d - r_f + 0.5 * sigma**2)*T) / (sigma * np.sqrt(T))
    d2 = d1 - sigma * np.sqrt(T)
    call_price = S * np.exp(-r_f * T) * norm.cdf(d1) - K * np.exp(-r_d * T) * norm.cdf(d2)
    return call_price

# 3. Build a strike grid around the forward (say +/- 10% around F0)
num_strikes = 50
K_vals = np.linspace(0.90*F0, 1.10*F0, num_strikes)

call_prices = []
for K in K_vals:
    # 3a. Numerically find the implied vol consistent with the delta-spline
    sigma_star = delta_to_sigma(delta_target=None, K=K)
    # 3b. Price the call with that implied vol
    if not np.isnan(sigma_star):
        call_prices.append(black_scholes_call(S0, K, r_dom, r_for, sigma_star, tau))
    else:
        call_prices.append(np.nan)

call_prices = np.array(call_prices)

# 4. Approximate RND via second difference
# f_RN(K) = e^{r_dom*T} * d^2C/dK^2  (depending on your convention).
# We'll do a simple finite difference:

f_rn = []
dK = K_vals[1] - K_vals[0]  # uniform spacing
discount_factor = np.exp(r_dom * tau)  # or check your sign/currency conventions

for i in range(1, len(K_vals)-1):
    d2C = (call_prices[i+1] - 2*call_prices[i] + call_prices[i-1]) / (dK**2)
    # RND at midpoint
    f_val = discount_factor * d2C
    f_rn.append(f_val)

K_mid = K_vals[1:-1]  # corresponding K's for the second difference

# Now f_rn array contains approximate risk-neutral pdf values at K_mid.
```

```python
import matplotlib.pyplot as plt

# K_mid and f_rn were computed in the code snippet above.

# 1) Plot in terms of the strike K:
plt.figure(figsize=(7,4))
plt.plot(K_mid, f_rn, marker='o', label='RND')
plt.title("Estimated Risk-Neutral PDF vs. Strike (K)")
plt.xlabel("Strike K")
plt.ylabel("f_RN(K)")
plt.grid(True)
plt.legend()
plt.show()

# 2) (Optional) Plot vs. moneyness, e.g. (K - F0)/F0:
moneyness_mid = (K_mid - F0)/F0
plt.figure(figsize=(7,4))
plt.plot(moneyness_mid, f_rn, marker='o', color='orange', label='RND')
plt.title("Estimated Risk-Neutral PDF vs. Moneyness ((K-F0)/F0)")
plt.xlabel("Moneyness")
plt.ylabel("f_RN(K)")
plt.grid(True)
plt.legend()
plt.show()
```
